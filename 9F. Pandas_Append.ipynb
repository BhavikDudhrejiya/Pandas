# APPEND

Append function is used to append rows of other data frame to the end of the given data frame, returning a new data frame object.

```python
#Importing library
import pandas as pd
```

```python
#Creating new data fra,e
df1 = pd.DataFrame({'A':[1,2,3],
                   'B':[10,20,30]},
                  index = ['a','b','c'])

df2 = pd.DataFrame({'A':[4,5,6],
                   'B':[40,50,60]})
#View the data frame
display(df1,df2)
```

```python
#One data frame appending to another data frame
df1.append(df2)
```

```python
#One data frame appending to another data frame with ignoring index
df1.append(df2, ignore_index = True)
```

```python
#Creating data frame
df1 = pd.DataFrame({'A':[1,2,3],
                   'B':[10,20,30]},
                  index = ['a','b','c'])

df2 = pd.DataFrame({'C':[4,5,6],
                   'B':[40,50,60]})

display(df1,df2)
```

```python
#one data frame appending to another data frame with ignoring index
df1.append(df2, ignore_index = True)
```
