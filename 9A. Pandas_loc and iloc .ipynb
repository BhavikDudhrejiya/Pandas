# PANDAS LOC

Access a group of rows and columns by labels or a boolean array.

```python
#Importing library
import pandas as pd
```

```python
data = pd.read_csv('C:\\Users\\bhavi\\OneDrive\\Desktop\\Data\\Student.csv')
data
```

```python
#Accessing the first row
data.loc[0]
```

```python
#Accessing the fifth row
data.loc[5]
```

```python
#Accessing the first, third and fifth row
data.loc[[0,2,4]]
```

```python
#Accessing from first to ninth row
data.loc[0:9]
```

```python
#Accessing the sepcific values of the given variable
data.loc[4,'Class']
```

```python
#Accessing the sepcific values of the given variable
data.loc[10,'Percentage']
```

```python
#Accessing the range values of the given variable
data.loc[0:10,'Percentage']
```

```python
#Accessing the row through boolean
data.loc[[False, False, False, True, True, False, False, True, True, True, False]]
```

```python
#Conditional Operator Using in loc
data.loc[data['Class']<11]
```

```python
#Conditional Operator Using in loc
data.loc[data['Class']<11, ['Percentage']]
```

# PANDAS ILOC

Integer location based indexing

```python
#Accessing the row through integer base indexing - showing vertical
data.iloc[0]
```

```python
#Accessing the row through integer base indexing - showing horizontal
data.iloc[[0]]
```

```python
#Accessing the single column
data.iloc[:,0]
```

```python
#Accessing the multiple column in a range
data.iloc[:,0:5]
```

```python
#Accessing the mutiple rows but not range 
data.iloc[[0,1]]
```

```python
data.iloc[[True,True,True,True,True,True,True,True,True,True,True]]
```
