```python
#Importing library
import pandas as pd
```

```python
#Reading data
data = pd.read_csv("C:\\Users\\bhavi\\OneDrive\\Desktop\\Data\\Movie_Ratings.csv")
data
```

## PIVOT TABLE WITH INDEX PARAMETER

```python
#Creating Pivot Table with index as Genre
data.pivot_table(index = 'Genre')
#We are getting mean value by default
```

```python
#Creating Pivot Table with index as Genre
data.pivot_table(index = 'Year')
```

## PIVOT TABLE WITH COLUMNS PARAMETER

```python
#Creating pivot table with columns
data.pivot_table(index = 'Genre', columns = 'Year')
```

## PIVOT TABLE WITH AGGFUNC PARAMETER

```python
#Creating pivot table using aggregate function
data.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'count')
```

## PIVOT TABLE WITH FILL_VALUE PARAMETER

```python
#Creating pivot table with filling values in NA
data.pivot_table(index = 'Genre', columns = 'Year', fill_value = 'none')
```

```python
#Creating pivot table with filling values in NA
data.pivot_table(index = 'Genre', columns = 'Year', fill_value = 0)
```

## PIVOT TABLE WITH MARGINS PARAMETER

```python
#Creating pivot table with adding margins
data.pivot_table(index = 'Genre', columns = 'Year', fill_value = 0, margins = True)
```
