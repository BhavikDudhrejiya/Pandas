# pd.read_csv

Important Parameters in read_csv
1. nrow :  Accessing the rows
2. usecols : Accessing the columns
3. skiprows : Skipping the unwanted rows above the header rows or skipping wanted rows
4. index_col : Creating any existing column as an index
5. header : Giving the index number of the header in the data
6. header = None : Giving None to the header if the data does not have the header so it will gives a index number
7. prefix : Giving the prefix like data/column/col etc when hears is none but must have header = None parameter
8. name : Giving name of the columns
