## Importing library

```python
#Importing Libraries
import pandas as pd
import numpy as np
```

## Reading the data

```python
#Loading data
data = pd.read_csv('C:/Users/bhavi/OneDrive/Desktop/Data/Data.csv')
data
```

## Text Value Replace

```python
#Replacing Pass into capital letter PASS
data.replace(to_replace = 'Pass', value = 'PASS')
#or 
#data.replace('Shipped','shipped')
```

## Numeric Value Replace

### Replacing Single Value

```python
#Replacing 1,2,3 values in the data by 0
data.replace(to_replace=80,
             value=85)
```

```python
#Replacing 1,2,3 values in the data by 0
data.replace(to_replace={'Class':[1,2,3]},
             value={'Class':['First Class','Second Class','Third Class']})
```

```python
#Replace Multiple values
data.replace([2,3],0)
```

```python
#Replacing 'Bhavik','Amit','Pooja' by 'Bhavik Dudhrejiya', 'Amit Shah', 'Pooja Bedi'
data.replace(['Bhavik','Amit','Pooja'],['Bhavik Dudhrejiya', 'Amit Shah', 'Pooja Bedi'])
```

```python
data.replace({'Result':'Pass'},'none')
```
