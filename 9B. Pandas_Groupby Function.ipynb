# GROUPBY

Pandas groupby function is used to split the data into groups based on some criteria

```python
#Importing library
import pandas as pd
```

```python
#Reading the data
data = pd.read_csv('C:\\Users\\bhavi\\OneDrive\\Desktop\\Data\\Student.csv')
data
```

```python
#Groupby data by Section
gr = data.groupby('Section')
gr.first()
```

```python
#Checking the groups
gr.groups
```

```python
#Groupby data by Class
gr1 = data.groupby('Class')
gr1.first()
```

```python
#Checking the groups
gr1.groups
```

```python
#Groupby data by Section and Class
gr3 = data.groupby(by = ['Section','Class'])
gr3.first()
```

```python
#Viewing the groupwise data
for Section, data in gr3:
    print(Section)
    print(data)
```

```python
#Viewing the groupby data in the form of list
list(gr3)
```

```python
#View the groupby data in the form of dictionary
dict(list(gr3))
```
