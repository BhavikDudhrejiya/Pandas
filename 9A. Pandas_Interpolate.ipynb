# INTERPOLATE FUNCTION WITH METHOD PARAMETER

What is interpolate function ?
Pandas interpolate function is basically used to fill nan values in dataframe or series

```python
#Importing library
import pandas as pd
```

```python
#Reading file
data = pd.read_csv('C:\\Users\\bhavi\\OneDrive\\Desktop\\Data\\Student1.csv')
display(data)
```

## Interpolating without paramter. Filling only numeric values

### Without method

```python
#Using interpolation function without any parameters
data.interpolate()
#It is  only guessing numeric value. It is not guessing string values.
#It will replace missing value = (Forward value of NA + Backward value of NA)/2
```

## Interpolating with linear method. Filling only numeric values

```python
#Interpolating with method linear. Filling only numeric values
data.interpolate(method = 'linear')
```

# INTERPOLATE FUNCTION WITH LIMIT_DIRECTION PARAMETER

```python
#Reading file.
data = pd.read_csv('C:\\Users\\bhavi\\OneDrive\\Desktop\\Data\\Student1.csv')
data
```

## Interpolating with limit_direction backward

```python
#Interpolating with limit_direction backward
data.interpolate(limit_direction = 'backward')
```

## Interpolating with limit_direction forward

```python
#Interpolating with limit_direction forward
data.interpolate(limit_direction = 'forward')
```

## Interpolating with limit_direction both

```python
#Interpolating with limit_direction both
data.interpolate(limit_direction = 'both')
```

# INTERPOLATE FUNCTION WITH LIMIT_AREA PARAMETER

## Interpolating with limit_area inside

```python
#Interpolating with limit area inside
data.interpolate(limit_area = 'inside')
```

# Interpolating with limit_area outside

```python
# Interpolating with limit area inside
data.interpolate(limit_area = 'outside')
#It will not change the data
```

# INTERPOLATE FUNCTION WITH INPLACE PARAMETER

## Interpolating with Inplace is True

```python
#Interpolating with Inplace is True
data.interpolate(inplace = True)
data
#It will change data in original data
```

## Interpolating with Inplace is False

```python
#Interpolating with Inplace is False
data.interpolate(inplace = False)
data
#It will not change data in original data
```
