# CONCAT FUNCTION

What is Pandas Concat Function?
Pandas concat function provide various facilities for easily combining together series, data frame and panel object.

## Concatinating Series

```python
#Importing library
import pandas as pd
```

```python
#Creating series
sr1 = pd.Series([0,1,2,3])
sr1
```

```python
sr2 = pd.Series([3,4,5,6])
sr2
```

```python
#Concating sr1 and sr2
pd.concat([sr1,sr2])
```

## Concatinating Data Frame row wise

```python
#Creating data frame df1
df1 = pd.DataFrame({'ID':[1,2,3,4],
                   'Name':['A','B','C','D'],
                   'Class':[5,6,7,8]})
df1
```

```python
#Creating data frame df2
df2 = pd.DataFrame({'ID':[5,6,7,8],
                   'Name':['E','F','G','H'],
                   'Class':[9,10,11,12]})
df2
```

```python
#Concatinating data frames df1 and df2
pd.concat([df1,df2])
```

```python
#Concatinating data frames df2 and df1
pd.concat([df2,df1])
```

```python
#Concatinating data frames df1 and df2
pd.concat([df1,df2], axis = 0)
```

# Concatinating Data Frame column wise

```python
#Concatinating data frames df1 and df2
pd.concat([df1,df2], axis = 1)
```

# Reindixing in the data after concat

```python
#Concatinating data frames df1 and df2
pd.concat([df1,df2], axis = 0, ignore_index=True)
```

```python
#Creating data frame df2
df3 = pd.DataFrame({'ID':[3,4],
                   'Name':['C','D'],
                   'Class':[7,8]})
df3
```

```python
#Column wise concat
pd.concat([df1,df3], axis = 1)
```

```python
#Row wise concat
pd.concat([df1,df3], axis = 0)
```

```python
#Column wise concat
pd.concat([df1,df3], axis = 1, join = 'inner')
```

```python
#Column wise concat
pd.concat([df1,df3], axis = 1, join = 'outer')
```

```python
#Concatinating two data frame with creating keys
pd.concat([df1,df2], keys = ['df1','df2'])
```

```python
#Concatinating two data frame with creating keys
pd.concat([df1,df2], keys = ['1st Data Frame','2nd Data Frame'])
```

```python
#Concatinating two data frame with creating keys
pd.concat([df1,df2],axis = 1, keys = ['1st Data Frame','2nd Data Frame'])
```

```python
#Creating data frame through dictionary
df4 = pd.DataFrame({'Marks': [40,63,91,34]})
df4
```

```python
#Concatinating data frame1 with data frame 4
pd.concat([df1,df4])
```

```python
#Sorting while concatinating (Ascendig order)
pd.concat([df1,df4], sort = True)
```

```python
#Sorting while concatinating(Discending order)
pd.concat([df1,df4], sort = False)
```
